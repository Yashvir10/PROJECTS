```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text = "The quick brown foxes are jumping over the lazy dogs! Running fast, they laughed."
```

```python
# text cleaning 
def clean_text(text):
    text = text.lower()                         
    text = re.sub(r'[^a-z\s]', '', text)          
    text = re.sub(r'\s+', ' ', text).strip()    
    return text
```

```python
cleaned_text = clean_text(text)
print(cleaned_text)
```

```python
# tokenization
tokens=word_tokenize(cleaned_text)
print("tokens:\n",tokens)
```

```python
# STOPWORDS REMOVAL
stop_words=set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token not in stop_words]    
print("filtered_tokens:\n",filtered_tokens)
```

```python
# stemming
stemmer=PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens]
print("stemmed_tokens:\n",stemmed_tokens)   
```

```python
# lemmatization
lemmatizer=WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token, pos='v') for token in filtered_tokens]
print("lemmatized_tokens:\n",lemmatized_tokens)
```

```python
corpus = ["the quick brown fox", "the lazy dog sleeps"]
vectorizer = CountVectorizer()
vectors = vectorizer.fit_transform(corpus)
```

```python
print("5. Vocabulary:", vectorizer.get_feature_names_out())
print("   Vectors:\n", vectors.toarray())
```
